```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
dctry='/home/morteza/dlc_projects/Analysis/Currencodes/TDA/Visualization/Lyps/'
```

```python
lyp_df=pd.read_csv(dctry+'Lyps_1.csv').dropna()
lyp_df.head()
```

```python

flatui = ['dimgray', 'darkgreen', 'blue', 'darkviolet']
sns.set_palette(sns.color_palette(flatui))

lyp_df['treatment'].replace(['SKF', 'D1A', 'D2A'], ['D1+', 'D1-','D2-'], inplace=True)
sns.lineplot(x='dim', y='log_lyp', hue='treatment', data=lyp_df.loc[lyp_df['treatment'].isin(['D1+','D2K', 'D2-','LID'])])



plt.xlim(2,33)

plt.xlabel('Dimensions', fontsize='large')
plt.ylabel('log Lypanov Exp.', fontsize='large')

plt.legend(frameon=False)

sns.despine()

plt.savefig('Lyps.svg')
```

```python
fig1 = plt.figure(figsize=(7, 7))
ax1=fig1.add_subplot(111)

flatui = ['silver', 'darkseagreen', 'steelblue', 'plum']
sns.set_palette(sns.color_palette(flatui))



kwrg={'width':0.3}

sns.barplot(x='treatment', y='lyp', data=lyp_df.loc[lyp_df['dim'].isin([15])], ax=ax1,
        order=['D1+','D2K', 'D2-','LID'])#, showfliers=False)
sns.despine()

plt.xlabel('' )
#plt.text(1, 0.24, '11 Dimensions',fontsize='large')
plt.ylabel('Max. Lyapanpov Exponent',fontsize='large')

plt.xlim(-1, 7)


plt.savefig('Lyp_bars.svg')
```

```python
fig1 = plt.figure(figsize=(7, 7))
ax1=fig1.add_subplot(111)

kwrg={'width':0.3}

sns.barplot(x='treatment', y='lyp', data=lyp_df.loc[lyp_df['dim'].isin(['21'])], ax=ax1,
        order=['LES','LID', 'LD6'], ci=95)#, showfliers=False)
sns.despine()
plt.title('7 Dimentiosn Included', loc='left')
plt.xlabel('Treatment Group', loc='left')
plt.ylabel('Max. Lyapanpov Exponent')

plt.xlim(-1, 7)
```

```python
from scipy import stats
import numpy as np

g1=np.log(lyp_df['lyp'].loc[lyp_df['treatment']=='LID'].loc[lyp_df['dim']==25])
g2=np.log(lyp_df['lyp'].loc[lyp_df['treatment']=='D2A'].loc[lyp_df['dim']==25])

print(stats.kruskal(g1,g2)[1])
print(stats.ttest_ind(g1,g2)[1])
```
