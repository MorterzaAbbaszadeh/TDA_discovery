```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

flatui = ["#9b59b6", "#3498db", '#a9a9a9', '#008000', '#b8860b', '#7b68ee']
sns.set_palette(sns.color_palette(flatui))


pth='/home/morteza/dlc_projects/Analysis/Currencodes/TDA/simplex_proj/12D/' #GET A LIST OF DATA
dset=[]


for root, dirs, files in os.walk(pth):
    for fname in files:         #8 third 
        if fname.endswith('csv') :
            case=root+'/'+fname #construct the path to file
            df=pd.read_csv(case).dropna()   #read Body_head
                
            for j in np.unique(df['dimension'].values):
                    
                dset.append((fname[12:15], fname[15:18], j , np.mean(df[df['dimension']==j]                                                                                        ['prediction_length'])))

simplex_df=pd.DataFrame(dset, columns=['treatment', 'ID', 'dimension', 'predicted_length'])
```

```python
pth='/home/morteza/dlc_projects/Analysis/Currencodes/TDA/simplex_proj/shuffle/' #GET A LIST OF DATA
dset=[]


for root, dirs, files in os.walk(pth):
    for fname in files:         #8 third 
        if fname.endswith('csv') :
            case=root+'/'+fname #construct the path to file
            df=pd.read_csv(case).dropna()   #read Body_head
                
            for j in np.unique(df['dimension'].values):
                    
                dset.append(('scrambled', fname[11:14], j , np.mean(df[df['dimension']==j]                                                                                        ['prediction_length'])))

shuffle_df=pd.DataFrame(dset, columns=['treatment', 'ID', 'dimension', 'predicted_length'])

```

```python
final_df=simplex_df.append(shuffle_df)


flatui = ["#9b59b6",'#a9a9a9']
sns.set_palette(sns.color_palette(flatui))


sns.lineplot(x='dimension', y='predicted_length', hue='treatment', data=final_df.loc[final_df['treatment'].isin(['SHM','LES', 'LID'])])

plt.legend(frameon=False)

plt.xlim(0,40)
plt.ylim(0,500)
plt.xlabel('Dimensions')
plt.ylabel('Predicted Length')
sns.despine()
plt.savefig('sham_prediction')
```

```python
from scipy import stats
import numpy as np

g1=np.array(final_df['predicted_length'].loc[(final_df['dimension']==6) & (final_df['treatment']=='SHM')])
g2=np.array(final_df['predicted_length'].loc[(final_df['dimension']==6) &( final_df['treatment']=='LES')])

print(stats.ks_2samp(g1,g2)[1])
print(stats.ttest_ind(g1,g2)[1])
```
