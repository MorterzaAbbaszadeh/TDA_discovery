```python
flatui = ['dimgray', 'darkgreen', 'darkviolet']
sns.set_palette(sns.color_palette(flatui)) 



sns.lineplot(x='dimension', y='size', hue='Embedding', data=df.loc[df['treatment'].isin(['SHM'])], ci=95)
plt.legend(frameon=False)
sns.despine()
plt.ylabel('Log(F.N.N.)')
plt.xlim(0,60)
plt.savefig('FNN_line.svg')
```

```python
flatui = ['silver', 'darkseagreen',  'plum']
sns.set_palette(sns.color_palette(flatui))

fig1 = plt.figure(figsize=(4, 5))
ax1=fig1.add_subplot(111)

df2=df.loc[df['treatment']=='SHM']
sns.boxplot(x='Embedding', y='size', data=df2.loc[df2['dimension'].isin([11])], ax=ax1,
        order=['body', 'head', 'Combined'], showfliers=False)
sns.despine()

ax=plt.gca()
ax.set_xticks([0,1,2])
ax.set_xticklabels(['Body','Head', 'Comb.'])
plt.ylabel('Log(F.N.N)', fontsize='large')
plt.xlabel('')
#plt.text(0, -1.5,'Dimension 11', fontsize='large')
plt.xlim(-.7, 5)

plt.savefig('FNN_box.svg')
```
